## Checking the Missing Values

#### Filling Numeric missing values

```python
# filling with median value
median_br = df.NUM_BEDROOMS.median()
print(median_br)
```

```python
# filling with median value
median_br = df.NUM_BEDROOMS.median()
print(median_br)
df.NUM_BEDROOMS.fillna(median_br)
```

```python
#df.info
df.info()
```

```python
# DF head
df.head(9)
```

```python
# filling with median value
median_br = df.NUM_BEDROOMS.median()
print(median_br)
df.NUM_BEDROOMS.fillna(median_br, inplace = True)
```

```python
# filling with mean
mean_nb = df.NUM_BATH.mean()
print(mean_nb)
```

```python
# filling with mean
mean_nb = df.NUM_BATH.mean()
print(mean_nb)
df.NUM_BATH.fillna(mean_nb, inplace = True)
```

```python
# df info
df.head(9)
```

```python
# filling with a specific number
df.ID.fillna(999999, inplace = True)
df.head(9)
```
