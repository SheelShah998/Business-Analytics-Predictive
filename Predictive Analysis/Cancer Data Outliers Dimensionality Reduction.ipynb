### Removing outliers greater than 3 sigma 

```python
# High value 
am_plus3sd = df.area_mean.mean() + 3 * df.area_mean.std()
am_plus3sd
```

```python
# Removing outliers
df1 = df.query('area_mean <= 1710')
df1.info()
```

## Removing outliers using IQR

```python
# IQR
t = df.area_mean.quantile(0.75) + 1.5 * (df.area_mean.quantile(0.75) - df.area_mean.quantile(0.25))
t
```

# Predictive analytics: Classification Model

```python
# changing the target to numeric
df.diagnosis = np.where(df.diagnosis == 'M', 1, 0 )
```

```python
# Choose predictors
X = df.iloc[:, 1:]
```

```python
# target variable
y = df['diagnosis']
```

```python
# importing KNN
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 11)
```

```python
# standardizing predictors
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
X_std = sc.transform(X)
```

```python
# Crossvalidation
from sklearn.model_selection import cross_val_score, StratifiedShuffleSplit
shuffle_split = StratifiedShuffleSplit(test_size = 0.2, n_splits = 10, random_state = 123)
scores = cross_val_score(model, X_std, y, cv = shuffle_split, scoring = 'f1')
print('CV', scores)
print('Mean CV', scores.mean())
```
